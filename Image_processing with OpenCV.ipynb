```python
import urllib.request
```

```python
image_url = "https://images-eu.ssl-images-amazon.com/images/I/51AKNcreMiL._SX342_SY445_QL70_ML2_.jpg"
image_name = "pm.jpg"
```

```python
urllib.request.urlretrieve(image_url, image_name)
```

```python
os.listdir(os.curdir)
```

```python
import matplotlib.pyplot as plt
```

```python
img_data = cv2.imread(image_name)
```

```python
plt.imshow(img_data)
```

```python
img_corrected = cv2.cvtColor(img_data, cv2.COLOR_BGR2RGB)
plt.imshow(img_corrected)
```

```python
from pylab import rcParams

rcParams['figure.figsize'] = 10,8
plt.axis('off')
plt.imshow(img_corrected)
```

```python
img_gray = cv2.cvtColor(img_data, cv2.COLOR_BGR2GRAY)

rcParams['figure.figsize'] = 10,8
plt.imshow(img_gray, cmap = 'gray')
plt.axis('off')
```

```python
rcParams['figure.figsize'] = 10,10

edge = cv2.Canny(img_corrected, threshold1 = 100, threshold2 = 200)
plt.imshow(edge, cmap='gray')
plt.xticks([])
plt.yticks([])
```

```python
rcParams['figure.figsize'] = 10,10

plt.hist(img_gray.ravel(), bins=256, range=(0,256))
plt.title('gray scale image histogram')
plt.xlabel('pixels intensity')
plt.ylabel('numner of pixel')
plt.show()
```

```python
rcParams['figure.figsize'] = 10,6

color = ('r','g','b')

for i,col in enumerate(color):
    histg = cv2.calcHist([img_data], [i], None, [256], [0,256])
    plt.plot(histg, color = col)
plt.show()
```
